```python
# friend_chatbot_streamlit.py
import streamlit as st
from datetime import datetime

st.set_page_config(page_title="Campus Helper Chatbot", page_icon="🏫")
st.title("🏫 Campus Helper Chatbot")

# Initialize chat history
if "history" not in st.session_state:
    st.session_state.history = []

# Function to generate bot response
def bot_reply(user_msg):
    msg = user_msg.lower()

    if "hi" in msg or "hey" in msg:
        return "Hey there! I'm your Campus Helper 🤖. How can I assist you today?"
    elif "canteen" in msg:
        return "The canteen serves meals from 8 AM to 7 PM. Don't forget the combo meal deals!"
    elif "event" in msg or "fest" in msg:
        return "College events are updated on the notice board every Monday. Check it out!"
    elif "hostel" in msg:
        return "Hostel queries? Contact the warden at warden@college.edu or visit the office."
    elif "assignment" in msg:
        return "Assignments are due via the LMS portal before deadlines. Late submissions are penalized."
    elif "exam" in msg:
        return "Exam schedules are published online. Make sure you check your roll number carefully."
    elif "attendance" in msg:
        return "Maintain at least 75% attendance to avoid penalties. Keep track weekly!"
    elif "library" in msg:
        return "Library timing: 9 AM - 8 PM. Don’t forget to return books on time!"
    elif "time" in msg:
        return f"Current time is {datetime.now().strftime('%H:%M:%S')} ⏰"
    elif "bye" in msg or "goodbye" in msg:
        return "Goodbye! Have a productive day on campus 🏫"
    else:
        return "Hmm… I’m not sure about that. Ask me about attendance, exams, library, canteen, or events."

# User input
user_text = st.text_input("You: ", "")

if user_text:
    reply = bot_reply(user_text)
    st.session_state.history.append(("You", user_text))
    st.session_state.history.append(("Bot", reply))

# Display chat
for speaker, message in st.session_state.history:
    if speaker == "You":
        st.markdown(f"*{speaker}:* {message}")
    else:
        st.markdown(f"*{speaker}:* {message}")
```
